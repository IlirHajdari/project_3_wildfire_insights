```python
import pandas as pd
from pathlib import Path
```

```python
firedata = Path("Resources/Fire_Data_USGS_2000_onwards.csv", encoding="ISO-8859-1")
```

```python
firedatadf = pd.read_csv(firedata)
```

```python
firedatadf.head()
```

```python
firedatadf.dtypes
```

```python
#rename columns
firedatadf = firedatadf.rename(columns={"OBJECTID":"ID", "FPA_ID":"Fire Program Analysis ID", 
                                                "NWCG_REPORTING_UNIT_NAME":"Reporting Group",
                                                "FIRE_NAME":"Fire Name",
                                                "FIRE_YEAR": "Year",
                                                "DISCOVERY_DATE":"Discovery Date",
                                                "NWCG_CAUSE_CLASSIFICATION":"Cause Class",
                                                "NWCG_GENERAL_CAUSE":"Cause Description",
                                                "NWCG_CAUSE_AGE_CATEGORY":"Caused by a Minor",
                                                "CONT_DATE":"Contained Date",
                                                "FIRE_SIZE":"Fire Size in Acres",
                                                "FIRE_SIZE_CLASS":"Fire Size Class",
                                                "LATITUDE":"Latitude",
                                                "LONGITUDE":"Longitude",
                                                "STATE":"State",
                                                "DISCOVERY_DOY":"Discovery Date (Days out of the Year)",
                                                "CONT_DOY":"Contained Date (Days out of the Year)",
                                                "FIPS_NAME":"County Name",
                                               "FIPS_CODE":"County Code"})
```

```python
#Reorder Columns
firedatadf = firedatadf[["ID", "Fire Program Analysis ID", "Latitude", "Longitude", "County Name", "County Code", "State", "Reporting Group", 
                                  "Fire Name", "Year", "Discovery Date", "Discovery Date (Days out of the Year)", 
                                  "Contained Date", "Contained Date (Days out of the Year)", "Cause Class", "Cause Description", "Caused by a Minor",
                                  "Fire Size in Acres", "Fire Size Class"]]

firedatadf.head()
```

```python
#checking for null count
nulls = firedatadf.isnull().sum()
print(nulls)
```

```python
firedatadf = firedatadf.fillna({"County Name":'Not Listed',"County Code":0, "Caused by a Minor":'No', "Fire Name":'Unnamed', 
                                        "Contained Date":'Unknown',"Contained Date (Days out of the Year)":0})
firedatadf.head()
```

```python
firedatadf["Caused by a Minor"] = firedatadf["Caused by a Minor"].replace("Minor", "Yes")
```

```python
#ninetiesto03df.head()
firedatadf.head()
```

```python
firedatadf["Contained Date (Days out of the Year)"] = firedatadf["Contained Date (Days out of the Year)"].astype(int)
firedatadf["County Code"]=firedatadf["County Code"].astype(int)
firedatadf.head()
```

```python
firedatadf["Discovery Date"] = pd.to_datetime(firedatadf["Discovery Date"])
```

```python
firedatadf.head()
```

```python
firedatadf["Days to Contain"] = firedatadf["Contained Date (Days out of the Year)"]-firedatadf["Discovery Date (Days out of the Year)"]
```

```python
firedatadf = firedatadf[["ID", "Fire Program Analysis ID", "Latitude", "Longitude", "County Name", "County Code", "State", "Reporting Group", 
                                  "Fire Name", "Year", "Discovery Date", "Discovery Date (Days out of the Year)", 
                                  "Contained Date", "Contained Date (Days out of the Year)", "Days to Contain", "Cause Class", "Cause Description", "Caused by a Minor",
                                  "Fire Size in Acres", "Fire Size Class"]]
firedatadf.head()
```

```python
firedatadf.to_json('cleanedfiredata.json',orient='records')
```
